```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df = df[["County","City","Model Year","Make","Electric Vehicle Type","Electric Range"]]
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.head()
```

```python
# bar plot for visualizing the categorical variable

plt.figure(figsize = (10,8))
sns.countplot(x = df["Electric Vehicle Type"])
plt.xlabel("Electric Vehicle Type")
plt.ylabel("Count of Vehicles")
plt.title("Type of Electric Vehicles")
plt.show()
```

```python
# histogram for visualizing the continuous variable

plt.figure(figsize = (10,8))
sns.histplot(x = df["Model Year"],binwidth=2,color='orange', edgecolor='black')
plt.xlabel("Model Year")
plt.ylabel("No.of Vehicles")
plt.show()
```
